### 41 データ読み込み

```python
import glob
import numpy as np
import pandas as pd
pd.set_option("max_rows", 500, "max_colwidth", 500, "max_columns", 500, "display.unicode.east_asian_width", True)
# google colabで使う場合は必須 => pip install -U plotly pandas
pd.options.plotting.backend = "plotly" # お好み
pd.options.display.float_format = "{:4f}".format
import plotly.express as px
```

```python
file_paths = glob.glob("./out_0001/*")
file_paths.sort()
file_paths[0:5]
```

```python
df_tmp = pd.read_csv(file_paths[0])
print(df_tmp.dtypes)
df_tmp.head()
```

```python
df = pd.concat([pd.read_csv(path, parse_dates=["receive_time"]) for path in file_paths])
df.head()
```

```python
print(df.shape)
print(df.dtypes)
print(df.isnull().sum())
```

# 41 

```python
date_min, date_max = df.receive_time.agg(["min", "max"])
print(date_min, "~", date_max)
print(date_max - date_min)
```

#42

```python
df["receive_date"] = df.receive_time.dt.date
df.head()
```

```python
df.groupby("receive_date").size().plot.bar()
```

#44

```python
df["dayofweek"] = df.receive_time.dt.dayofweek
df["day_name"] = df.receive_time.dt.day_name()
df.head()
```

```python
df[["receive_date", "dayofweek", "day_name"]].drop_duplicates(subset="receive_date")
```

```python
import datetime as dt
```

```python
df_ex = df.loc[(dt.datetime(2021, 1, 21) <= df.receive_time) & (df.receive_time < dt.datetime(2021, 1, 23))].copy()
print(df_ex.shape)
df_ex.head()
```

```python
df_ex["receive_time_sec"] = df_ex.receive_time.dt.floor("S")
print(df_ex.shape[0], df_ex.receive_time_sec.unique().shape[0])
```

```python
df_ex[df_ex.receive_time_sec.duplicated(keep=False)]
```

```python
print(df_ex.shape)
df_ex.drop_duplicates(subset="receive_time_sec", inplace=True)
print(df_ex.shape)
```
