```python
#5
tmp = "I am an NLPer"
def n_gram(text, n_gram_range=1, is_char=False):
  if is_char:
    word_list = text
  else:
    word_list = tmp.split(" ")
  return [tuple(word_list[idx: idx+n_gram_range]) for idx in range(0, len(word_list) - n_gram_range + 1)]

print(n_gram(tmp, n_gram_range = 2, is_char=True))
print(n_gram(tmp, n_gram_range = 2, is_char=False))
```

```python
#6
tmp1, tmp2 = "paraparaparadise", "paragraph"
X = set(n_gram(tmp1, n_gram_range=2, is_char=True))
Y = set(n_gram(tmp2, n_gram_range=2, is_char=True))
print(X, Y)
print("和集合", X | Y)
print("積集合", X & Y)
print("差集合", X - Y)
print("差集合", Y - X)
print("X in 'se'", ("s", "e") in X)
print("Y in 'se'", ("s", "e") in Y)
```

```python
#8 
min_ord = ord("a")
max_ord = ord("z")
def chiper(text):
  return "".join([chr(219 - ord(char)) if (min_ord <= ord(char)) & (ord(char) <= max_ord) else char for char in text])
chiper("I am an NLPer")
```

```python
#9
import random
tmp = "I couldn’t believe that I could actually understand what I was reading : the phenomenal power of the human mind ."
rt_list = []
for word in tmp.split(" "):
  # 4文字以下なら何も処理しない
  if len(word) <= 4:
    continue

  chars_list = list(word[1:-1])
  random.shuffle(chars_list)
  rt_list.append(word[0] + "".join(chars_list) + word[-1])
rt_list
```

```python
#20
england_text = ""
with open("./jawiki-country.json", "r") as f:
  for line in f.readlines():
    data = json.loads(line)
    if data["title"] == "イギリス":
      england_text = data["text"]
      break

england_text
```

```python
print(england_text)
```

```python
#21
extract_category_rows = [row for row in england_text.split("\n") if "Category" in row]
extract_category_rows
```

```python
import re
#22
pattern = re.compile(r"\[\[Category:(.+)\]\]")
for row in extract_category_rows:
  m = pattern.match(row)
  if m is not None:
    print(m.groups()[0])
```

```python
pattern = re.compile(r"^(\={2,})\s*(.*)\s*(\={2,}).*")
rt_dict = dict()
for row in england_text.split("\n"):
  m = pattern.match(row)
  if m is not None:
    equals, val, _ = m.groups()
    rt_dict[val.strip("=")] = len(equals) - 1
rt_dict
```

```python
#24
pattern = re.compile(r"\[\[ファイル:(.+?)\|")
for row in england_text.split("\n"):
  m = pattern.match(row)
  if m is not None:
    print(m.groups()[0])
```
